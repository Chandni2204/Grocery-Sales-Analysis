```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
data = pd.read_csv('/content/SuperMarket Analysis.csv')
```

```python
# Convert Date to datetime
data['Date'] = pd.to_datetime(data['Date'])
```

```python
# Example: Total sales per product line
product_sales = data.groupby('Product line')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,5))
sns.barplot(x=product_sales.index, y=product_sales.values)
plt.xticks(rotation=45)
plt.title('Total Sales per Product Line')
plt.show()
```

```python
# Example: Sales by Branch
branch_sales = data.groupby('Branch')['Sales'].sum()
plt.figure(figsize=(8,5))
sns.barplot(x=branch_sales.index, y=branch_sales.values)
plt.title('Total Sales per Branch')
plt.show()
```

```python
# Example: Monthly Sales Trend
monthly_sales = data.groupby(data['Date'].dt.month)['Sales'].sum()
plt.figure(figsize=(10,5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```
